```python
from transformers import pipeline

# Initialize a text generation pipeline
generator = pipeline('text-generation', model='gpt2')

# 1. IMPLICIT vs EXPLICIT PROMPTS
print('--- 1. Implicit vs Explicit ---')
implicit_prompt = 'AI is'
explicit_prompt = 'Explain the concept of AI to a 5-year-old in one sentence:'

print(f'Implicit Output: {generator(implicit_prompt, max_new_tokens=20, truncation=True)[0]["generated_text"]}')
print(f'Explicit Output: {generator(explicit_prompt, max_new_tokens=30, truncation=True)[0]["generated_text"]}')

# 2. ZERO-SHOT PROMPTING (Using a dedicated zero-shot pipeline)
print('\n--- 2. Zero-Shot Classification ---')
classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
sequence = 'The new rocket engine passed its first vacuum test today.'
labels = ['politics', 'space', 'cooking']

result = classifier(sequence, candidate_labels=labels)
print(f'Text: {sequence}')
print(f'Top Label: {result["labels"][0]} (Score: {result["scores"][0]:.4f})')

# 3. FEW-SHOT PROMPTING
print('\n--- 3. Few-Shot Text Generation ---')
few_shot_prompt = """Review: This movie was great! | Sentiment: Positive
Review: I hated the plot and the acting was bad. | Sentiment: Negative
Review: It was okay, nothing special. | Sentiment:"""

output = generator(few_shot_prompt, max_new_tokens=5, truncation=True)
print(f'Prompt with examples:\n{few_shot_prompt}')
print(f'Model Completion: {output[0]["generated_text"].split("Sentiment:")[-1].strip()}')
```
